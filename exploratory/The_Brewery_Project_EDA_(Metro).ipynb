```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://worldpopulationreview.com/us-cities"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
metroCitiesGroup = soup.find_all('table', class_='tp-table-body is-narrow w-full min-w-full table-auto border-separate border-spacing-0 border bg-white')
metroCitiesGroup = metroCitiesGroup[1]
print(metroCitiesGroup)
metroCities = metroCitiesGroup.find_all("tr",class_="odd:bg-white even:bg-gray-100")
#print(metroCities[10])
```

```python
cityNames = []
ranks = []
stateNames = []
totalPopulations = []
```

```python
for city in metroCities:
    cityName = city.find('a').get_text()
    cityDetails = city.find_all('td', class_="z-40 border px-2 py-0.5")
    cityRank = cityDetails[0].get_text()
    cityPopulation = cityDetails[1].get_text()
    cityStateTxt = city.find('a')['href'].split("-population")[0][-2:]
    if cityName and cityRank:
        cityNames.append(cityName)
        ranks.append(cityRank)
        stateNames.append(cityStateTxt)
        totalPopulations.append(cityPopulation)

dict = {'City Names': cityNames, 'Rank': ranks,"State": stateNames, '2024 Population': totalPopulations}
metroData = pd.DataFrame(dict)
metroData.head(10)
```

```python
#Fix Data Types:
#Particulatly the population column to be integer
for i in range(0,len(metroData["2024 Population"])):
    popNumeric = int(metroData["2024 Population"][i].replace(",",""))
    metroData["2024 Population"][i] = popNumeric
    metroData["Rank"][i] = int(metroData["Rank"][i])

```

```python
#Fix Data:
#Capitalize States
for i in range(0,len(metroData["State"])):
    stateCaps = metroData["State"][i].upper()
    metroData["State"][i] = stateCaps
```

```python
#Merge in region data to metroData
regions = pd.read_csv('https://raw.githubusercontent.com/cphalpert/census-regions/master/us%20census%20bureau%20regions%20and%20divisions.csv')
regions = pd.DataFrame(regions)
regions
```

```python
metroData = metroData.merge(regions, left_on='State', right_on='State Code')
metroData = metroData.sort_values(by = ["Rank"])
metroData = metroData.drop(["State Code"], axis = 1)
metroData = metroData.rename(columns={"State_x": "State Code", "State_y": "State Name"})

metroData.head(10)
```

```python
#Export metroData as .csv
metroData.to_csv('metropolianCities.csv')
```

##Exploratory Analysis

The Metropolitan Cities data set identifies the top 300 most populated cities in the United States. If breweries thrive in large cities, this will help us identify key examples.

This is a pretty large data set thought. Let's explore more about these cities and perhaps find a natural level to scope down the data set.

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

First, let's gather an understanding of how the populations of the top 300 cities are distributed. We can run some basic descriptive statistics and a plot to visualize the spread.

```python
mean = metroData["2024 Population"].mean()
min = metroData["2024 Population"].min()
median = metroData["2024 Population"].median()
max = metroData["2024 Population"].max()
sd = metroData["2024 Population"].std()

print("The mean of the populations is", round(mean,2),".")
print("The minimum of the populations is", round(min,2),".")
print("The median of the populations is", int(median),".")
print("The maximum of the populations is", round(max,2),".")
print("The standard deviation of the populations is", round(sd,2),".")
```

Reviewing these values, there is certainly an underlying skew to that data. There is a difference of approximately 70,00 between the lower 50% of these city populations. However, between the upper 50% of city populations, there is a difference of 7.7 million. That is a very noteworthy spread. Let's review this visually and against the Regions to understand if these larger cities concentrate anywhere in specific in the country.

```python
metroData = metroData.sort_values(by = "Region")
fig = px.box(metroData,x="2024 Population", color="Region",title="Distribution of Metropolitan Population by US Region",
            width = 800, height = 500)
fig.show()
```

We can see that there are several key cities that are extra large that are represented as outliers in the chart above. Each of the 4 US regions appears to exhibit this behavior with most major cities staying below approximately 500,000 people or fewer and a few extraneous mega-cities as well.

It is important to note, however from this visual that it is difficult to tell if there is a difference between regions about how many of these top 300 cities they host. Even if we included individual data point most of the lower 75% of data would be overlapping.

Let's review count of major cities by region for the full 300 cities, then top 150, then the top 50.

```python
top10 = metroData[metroData["Rank"]<=10]
top10Counts = top10['Region'].value_counts().sort_index()
print(top10)
top50 = metroData[metroData['Rank']<=50]
top50Counts = top50['Region'].value_counts().sort_index()
metroDataCounts = metroData['Region'].value_counts().sort_index()



labels = ["Midwest","Northeast","South","West"]

# Create subplots: use 'domain' type for Pie subplot
fig = make_subplots(rows=1, cols=3, specs=[[{'type':'domain'},{'type':'domain'}, {'type':'domain'}]],
                    subplot_titles = ["Top 300","Top 50","Top 10"])
fig.add_trace(go.Pie(labels=labels, values=metroDataCounts[0:4], name="Top 300"),
              1, 1)
fig.add_trace(go.Pie(labels=labels, values=top50Counts[0:4], name="Top 50"),
              1, 2)
fig.add_trace(go.Pie(labels=labels, values=top10Counts[0:4], name="Top 10"),
              1, 3)

# Use `hole` to create a donut-like pie chart
fig.update_traces(hole=.4, hoverinfo="label+percent+name+value")

fig.update_layout(
    title_text="Proportion of Metropolitan Cities by Region: Top 300, Top 50, and Top 10")
fig.show()
```

From this visual, we can identify key trends between the number of major cities each region hosts. It is clear that the majority about 80% of all major cities are held in the West or the South. Comparatively, the Midwest and the Northeast hold much fewer metropolitan cities. Although numbers shift some as we scope into the top 50 and the top 10, these proportion stay relatively consistent between the top 300 and the top 10.
